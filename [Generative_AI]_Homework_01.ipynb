# 繪製函數圖形

## 繪製 $f(x)=sin(x)$

```python
x = np.linspace(0, 2 * np.pi, 100) # x 範圍 0 到 2π 取樣 100 個點
y = np.sin(x)

plt.plot(x, y, linewidth=3) # 繪製座標點和線寬設定

plt.xlabel("x")
plt.ylabel("sin(x)")

plt.grid(True)
```

# 繪製特殊圖案

## 螺旋狀

```python
theta = np.linspace(0, 10*np.pi, 500)

r = theta
x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure(figsize=(8, 8)) # 設置圖的大小
plt.plot(x, y)

plt.grid(True)
```

## 蝴蝶

```python
plt.figure(figsize=(8, 8))

t = np.linspace(0, 12*np.pi, 3000)
r = np.exp(np.cos(t)) - 2*np.cos(4*t) + np.sin(t/12)**5
x = np.sin(t) * r
y = np.cos(t) * r

# 顏色用成漸層的
colors = np.linspace(0, 1, len(t))
plt.scatter(x, y, c=colors, s=1, cmap='hsv', alpha=0.8)
plt.axis('equal')
plt.grid(alpha=0.3)
```
